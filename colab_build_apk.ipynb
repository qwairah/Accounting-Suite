```python
# الخطوة 2: رفع ملفات المشروع
from google.colab import files
import os, zipfile, shutil

print('📁 يرجى رفع ملف ZIP يحتوي على: main.py + buildozer.spec')
uploaded = files.upload()

for fname, data in uploaded.items():
    with open(fname, 'wb') as f:
        f.write(data)
    if fname.endswith('.zip'):
        with zipfile.ZipFile(fname, 'r') as z:
            z.extractall('/content/qawera/')
        print(f'✅ تم استخراج {fname}')
    else:
        shutil.copy(fname, '/content/qawera/')

os.makedirs('/content/qawera', exist_ok=True)
os.chdir('/content/qawera')
print('📂 المجلد الحالي:', os.getcwd())
print('📋 الملفات:', os.listdir('.'))
```

```python
# الخطوة 3: إنشاء buildozer.spec إذا لم يكن موجوداً
if not os.path.exists('buildozer.spec'):
    spec_content = """[app]
title = قـويـره سوفت
package.name = qawerasoft
package.domain = com.qawera.soft
source.dir = .
source.include_exts = py,png,jpg,kv,atlas,json,db,ttf,txt
version = 2.0
requirements = python3,kivy==2.2.1,pillow,reportlab,sqlite3
orientation = portrait
android.permissions = INTERNET,WRITE_EXTERNAL_STORAGE,READ_EXTERNAL_STORAGE,READ_CONTACTS,CAMERA
android.api = 33
android.minapi = 21
android.ndk = 25b
android.accept_sdk_license = True
android.arch = arm64-v8a

[buildozer]
log_level = 2
"""
    with open('buildozer.spec', 'w') as f:
        f.write(spec_content)
    print('✅ تم إنشاء buildozer.spec')
else:
    print('✅ buildozer.spec موجود')
```

```python
# الخطوة 4: بناء APK
import os
os.chdir('/content/qawera')

print('🏗️ جاري بناء APK... قد يستغرق 15-30 دقيقة')
!buildozer -v android debug
print('✅ انتهى البناء!')
```

```python
# الخطوة 5: تحميل APK
from google.colab import files
import glob

apk_files = glob.glob('/content/qawera/bin/*.apk')
if apk_files:
    apk = apk_files[0]
    print(f'📱 APK جاهز: {apk}')
    files.download(apk)
    print('✅ تم تحميل APK بنجاح!')
else:
    print('❌ لم يتم العثور على APK - تحقق من الأخطاء أعلاه')
```
